**Fibonacci**

```python
def fib1(n):
    if n < 2:
        return n
    else:
        return fib1(n-1) + fib1(n-2)
```

```python
fib2 = lambda n: n if n < 2 else fib2(n-1) + fib2(n-2)
```

```python
print(fib1(10))
```

```python
import sympy as sp
import numpy as np

# Declarando variável simbólica
x = sp.Symbol('x')

# Função simbólica
f = x

# Derivada primeira
df = sp.diff(f, x)

# Derivada segunda
d2f = sp.diff(f, x, 2)

print("f(x) =", f)
print("f'(x) =", df)
print("f''(x) =", d2f)

# Integral definida de 0 a pi
integral = sp.integrate(f)
print(integral)
```
